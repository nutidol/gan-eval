```python
import pandas as pd
df = pd.read_csv('GAN_user_response.csv')


df.columns = ['timestamp', 'real_1', 'div_1', 'real_2', 'div_2', 'real_3', 'div_3', 'real_4', 'div_4',
             'real_5', 'div_5', 'real_6', 'div_6', 'real_7', 'div_7', 'real_8', 'div_8',
             'real_9', 'div_9', 'real_10', 'div_10', 'real_11', 'div_11', 'real_12', 'div_12',
             'comments', 'knowGAN']

df = df.dropna().reset_index(drop=True)
df.drop('timestamp', axis=1, inplace=True)

df
```

```python
SG2_all_real = (df['real_1'].value_counts().SG2 + df['real_2'].value_counts().SG2 + df['real_5'].value_counts().SG2 +
                df['real_6'].value_counts().SG2 + df['real_7'].value_counts().SG2 + df['real_8'].value_counts().SG2 +
                df['real_9'].value_counts().SG2 + df['real_11'].value_counts().SG2)


SG3_all_real = (df['real_1'].value_counts().SG3 + df['real_3'].value_counts().SG3 + df['real_4'].value_counts().SG3 +
                df['real_6'].value_counts().SG3 + df['real_7'].value_counts().SG3 + df['real_9'].value_counts().SG3 +
                df['real_10'].value_counts().SG3 + df['real_12'].value_counts().SG3)


SGXL_all_real = (df['real_2'].value_counts().SGXL + df['real_3'].value_counts().SGXL + df['real_4'].value_counts().SGXL + 
                 df['real_5'].value_counts().SGXL + df['real_8'].value_counts().SGXL + df['real_10'].value_counts().SGXL + 
                 df['real_11'].value_counts().SGXL + df['real_12'].value_counts().SGXL)

res = SG2_all_real+SG3_all_real+SGXL_all_real
SG2_real_avg = SG2_all_real/res
SG3_real_avg = SG3_all_real/res
SGXL_real_avg = SGXL_all_real/res

print('SG2_real', SG2_all_real, round(SG2_real_avg*100, 2))
print('SG3_real', SG3_all_real, round(SG3_real_avg*100, 2))
print('SGXL_real', SGXL_all_real, round(SGXL_real_avg*100, 2))
```

```python
SG2_all_div = (df['div_1'].value_counts().SG2 + df['div_2'].value_counts().SG2 + df['div_5'].value_counts().SG2 + 
                df['div_6'].value_counts().SG2 + df['div_7'].value_counts().SG2 + df['div_8'].value_counts().SG2 + 
                df['div_9'].value_counts().SG2 + df['div_11'].value_counts().SG2)
        
SG3_all_div = (df['div_1'].value_counts().SG3 + df['div_3'].value_counts().SG3 + df['div_4'].value_counts().SG3 + 
                df['div_6'].value_counts().SG3 + df['div_7'].value_counts().SG3 + df['div_9'].value_counts().SG3 + 
                df['div_10'].value_counts().SG3 + df['div_12'].value_counts().SG3)
        
SGXL_all_div = (df['div_2'].value_counts().SGXL + df['div_3'].value_counts().SGXL + df['div_4'].value_counts().SGXL + 
                df['div_5'].value_counts().SGXL + df['div_8'].value_counts().SGXL + df['div_10'].value_counts().SGXL +
                df['div_11'].value_counts().SGXL + df['div_12'].value_counts().SGXL)

res = SG2_all_div+SG3_all_div+SGXL_all_div
SG2_div_avg = SG2_all_div/res
SG3_div_avg = SG3_all_div/res
SGXL_div_avg = SGXL_all_div/res

print('SG2_div', SG2_all_div,  round(SG2_div_avg*100, 2))
print('SG3_div', SG3_all_div, round(SG3_div_avg*100, 2))
print('SGXL_div', SGXL_all_div, round(SGXL_div_avg*100, 2))
```
